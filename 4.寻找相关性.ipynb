```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
plt.style.use('ggplot')
warnings.filterwarnings(action="ignore")
boston=load_boston()###数据描述https://scikit-learn.org/stable/datasets/index.html#boston-dataset，数据两个属性data和target
data=pd.DataFrame(boston.data)
data['target']=pd.DataFrame(boston.target)
data.head()
```

```python
data.describe()#展现数值属性
```

```python
corr=data.corr()###数据之间的相关性
corr
```

```python
plt.subplots(figsize=(7,8))
sns.heatmap(corr,cmap="Blues",square=True)##square=True 整个图形为正方形
```

```python
corr['target'].sort_values(ascending=True)#####查看相关性
```
